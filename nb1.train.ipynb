```python
from sklearn.preprocessing import LabelEncoder

# Step 1: Identify categorical columns
categorical_cols = df.select_dtypes(include='object').columns.tolist()

# Step 2: Initialize label encoder
le = LabelEncoder()

# Step 3: Apply label encoding to each categorical column
for col in categorical_cols:
    df[col] = df[col].astype(str)  # Convert to string if needed
    df[col] = le.fit_transform(df[col])
```

Train/Validation Split

```python
from sklearn.model_selection import train_test_split

X = df.drop('IncidentGrade', axis=1)  # features
y = df['IncidentGrade']               # target (already encoded as integers)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, confusion_matrix, f1_score, precision_score, recall_score

# Initialize and train the model
model = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', random_state=42)
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_test)

# Evaluation Report
print("Classification Report:")
print(classification_report(y_test, y_pred, digits=4))

# Macro Metrics
macro_f1 = f1_score(y_test, y_pred, average='macro')
macro_precision = precision_score(y_test, y_pred, average='macro')
macro_recall = recall_score(y_test, y_pred, average='macro')

print(f"\nMacro F1 Score: {macro_f1:.4f}")
print(f"Macro Precision: {macro_precision:.4f}")
print(f"Macro Recall: {macro_recall:.4f}")

# Confusion Matrix
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))



```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, f1_score, precision_score, recall_score, confusion_matrix
from xgboost import XGBClassifier

# Step 1: Load data
df = pd.read_csv(r"C:\Users\VINOTH\Downloads\GUIDE_Train.csv\GUIDE_Train.csv")

# Step 2: Drop rows with missing target
df = df[df['IncidentGrade'].notna()]

# Step 3: Encode target
le = LabelEncoder()
df['IncidentGrade'] = le.fit_transform(df['IncidentGrade'])

# Step 4: Encode categorical features
categorical_cols = df.select_dtypes(include='object').columns
for col in categorical_cols:
    df[col] = df[col].astype(str)
    df[col] = LabelEncoder().fit_transform(df[col])

# Step 5: Drop unnecessary ID columns (optional)
drop_cols = ['Id', 'IncidentId', 'AlertId', 'OrgId']
df.drop(columns=[col for col in drop_cols if col in df.columns], inplace=True)

# Step 6: Sample a smaller subset of data (e.g., 30%)
df = df.sample(frac=0.3, random_state=42)

# Step 7: Split into X and y
X = df.drop('IncidentGrade', axis=1)
y = df['IncidentGrade']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Step 8: Define XGBoost and RandomizedSearchCV
param_dist = {
    'n_estimators': [100, 200],
    'max_depth': [4, 6],
    'learning_rate': [0.1],
    'subsample': [0.8, 1.0]
}

xgb = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', random_state=42)

random_search = RandomizedSearchCV(
    estimator=xgb,
    param_distributions=param_dist,
    n_iter=5,
    scoring='f1_macro',
    cv=3,
    verbose=1,
    n_jobs=1  # Run one job at a time to prevent memory overload
)

# Step 9: Fit model
random_search.fit(X_train, y_train)

# Step 10: Evaluate
best_model = random_search.best_estimator_
y_pred = best_model.predict(X_test)

print("Best Parameters:", random_search.best_params_)
print("\nClassification Report:")
print(classification_report(y_test, y_pred, digits=4))
print("Macro F1 Score:", f1_score(y_test, y_pred, average='macro'))
print("Macro Precision:", precision_score(y_test, y_pred, average='macro'))
print("Macro Recall:", recall_score(y_test, y_pred, average='macro'))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```
