```python
df.isnull().sum()
df.dropna(thresh=0.9 * len(df), axis=1, inplace=True)  # drop columns with >90% NaNs
```

```python
df.isnull().sum()
```

```python
df = df[df['IncidentGrade'].notnull()]
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder

# Step 1: Identify categorical columns
categorical_cols = df.select_dtypes(include='object').columns.tolist()

# Step 2: Initialize label encoder
le = LabelEncoder()

# Step 3: Apply label encoding to each categorical column
for col in categorical_cols:
    df[col] = df[col].astype(str)  # Convert to string if needed
    df[col] = le.fit_transform(df[col])
```

```python
import pickle

with open("label_encoder.pkl", "wb") as f:
    pickle.dump(le, f)
```

Train/Validation Split

```python
from sklearn.model_selection import train_test_split

X = df.drop('IncidentGrade', axis=1)  # features
y = df['IncidentGrade']               # target (already encoded as integers)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
from xgboost import XGBClassifier
from sklearn.feature_selection import SelectFromModel

# Train initial model
model = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', random_state=42)
model.fit(X_train, y_train)

# Feature selection
selector = SelectFromModel(estimator=model, threshold='median', prefit=True)

# Get boolean mask of selected features
mask = selector.get_support()

# List of all feature names
all_features = X_train.columns

# Get selected and removed features
selected_features = all_features[mask]
removed_features = all_features[~mask]

print("✅ Selected Features:")
print(selected_features.tolist())

print("\n❌ Removed Features:")
print(removed_features.tolist())
```

#MODEL RUNNING


```python
from xgboost import XGBClassifier
from sklearn.feature_selection import SelectFromModel
from sklearn.metrics import classification_report, confusion_matrix, f1_score, precision_score, recall_score

# Step 1: Train initial model
initial_model = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', random_state=42)
initial_model.fit(X_train, y_train)

# Step 2: Select important features
selector = SelectFromModel(estimator=initial_model, threshold='median', prefit=True)

# Step 3: Get selected feature names (optional, for display)
selected_feature_mask = selector.get_support()
selected_feature_names = X_train.columns[selected_feature_mask]
print("✅ Selected Features:")
print(selected_feature_names.tolist())

# Step 4: Transform datasets
X_train_selected = selector.transform(X_train)
X_test_selected = selector.transform(X_test)

# Step 5: Train model on selected features
model = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', random_state=42)
model.fit(X_train_selected, y_train)

# Step 6: Predict
y_pred = model.predict(X_test_selected)

# Step 7: Evaluation
print("\nClassification Report:")
print(classification_report(y_test, y_pred, digits=4))

macro_f1 = f1_score(y_test, y_pred, average='macro')
macro_precision = precision_score(y_test, y_pred, average='macro')
macro_recall = recall_score(y_test, y_pred, average='macro')

print(f"\nMacro F1 Score: {macro_f1:.4f}")
print(f"Macro Precision: {macro_precision:.4f}")
print(f"Macro Recall: {macro_recall:.4f}")

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
import pickle

# Save model to a file
with open("xgb_model.pkl", "wb") as file:
    pickle.dump(model, file)

print("Model saved as xgb_model.pkl")
```
